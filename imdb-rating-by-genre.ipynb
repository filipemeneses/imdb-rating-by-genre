```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set ipython's max row display
pd.set_option('display.max_row', 5)

# Set iPython's max column width to 50
pd.set_option('display.max_columns', 50)
```

```python
titles=pd.read_csv('title.basics.tsv', sep='\t')
ratings=pd.read_csv('title.ratings.tsv', sep='\t')
titlesWithRatings = titles.merge(ratings, left_on='tconst', right_on='tconst')
```

```python
genresInDataset = sorted(["Documentary","Short","Animation","Comedy","Romance","Sport","News","Fantasy","Horror","Drama","War","Western","Family","Adventure","Crime","Action","Music","Mystery","Biography","History","Sci-Fi","Musical","Thriller","Film-Noir","Game-Show","Talk-Show","Reality-TV","Adult"])
titlesByGenre = {}
for genre in genresInDataset:
    titlesByGenre[genre] = titlesWithRatings[titlesWithRatings.genres.str.contains(genre, na=False)]
```

```python
for genre in genresInDataset:
    titlesByGenre[genre].hist(column='averageRating')
    plt.title(genre)
```
